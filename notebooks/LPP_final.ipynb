## Data Preprocessing

Raw datasets often contain missing values, irrelevant columns, or categorical data that cannot be directly used by machine learning algorithms.

- Handling missing values
- Removing irrelevant columns such as image links

## Handling Missing Values

Some columns in the dataset contain missing values. Missing values can negatively affect the performance of machine learning models.

To handle missing data, numerical columns such as ratings, number of ratings, and number of reviews are filled using the median value of their respective columns. This ensures that the dataset remains consistent and usable for model training.

```python
df.isna().sum()
```

```python
df["rating"] = df["rating"].fillna(df["rating"].median())
df["no_of_ratings"] = df["no_of_ratings"].fillna(df["no_of_ratings"].median())
df["no_of_reviews"] = df["no_of_reviews"].fillna(df["no_of_reviews"].median())
```

```python
df.isna().sum()
```

```python
df = df.drop(["Unnamed: 0", "img_link"], axis=1, errors="ignore")
```

## Encoding Categorical Variables

Machine learning algorithms require numerical input data. However, many columns in the dataset contain categorical values such as processor type, RAM configuration, storage type, and operating system.

To make these features usable for machine learning models, categorical variables are converted into numerical values using encoding techniques such as label encoding.

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["name"] = le.fit_transform(df["name"])
df["processor"] = le.fit_transform(df["processor"])
df["ram"] = le.fit_transform(df["ram"])
df["os"] = le.fit_transform(df["os"])
df["storage"] = le.fit_transform(df["storage"])
df.head()
```

## Feature Selection

Feature selection involves identifying the input variables that will be used to train the machine learning model.

In this project, laptop specifications such as processor type, RAM, storage, operating system, display size, ratings, and reviews are used as input features.

The laptop price is selected as the target variable. The model learns to predict the laptop price based on the given specifications.

```python
X = df.drop("price(in Rs.)", axis=1)
Y = df["price(in Rs.)"]
X
```

## Splitting the Dataset

The dataset is divided into two parts:

Training Data – used to train the machine learning model  
Testing Data – used to evaluate the model performance

This approach helps measure how well the trained model performs on unseen data.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

## Model Training

Machine learning models are trained using the training dataset. The models learn patterns and relationships between laptop specifications and their corresponding prices.

In this project, regression algorithms are used because the target variable (price) is a continuous numerical value.

Different models such as Linear Regression, Decision Tree Regression, and Random Forest Regression can be used to train the dataset and learn price prediction patterns.

## Model Evaluation

After training the machine learning models, their performance is evaluated using the testing dataset.

Evaluation metrics such as the R-squared score help measure how accurately the model predicts laptop prices. A higher score indicates better prediction performance.

####Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
lr_pred=lr.predict(X_test)
from sklearn.metrics import r2_score
lr_score=lr.score(X_test,Y_test)
lr_score
```

####Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(X_train, Y_train)
```

```python
dt_pred=dt.predict(X_test)
dt_score=dt.score(X_test,Y_test)
dt_score
```

####Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train, Y_train)
```

```python
rf_pred=rf.predict(X_test)
rf_score=rf.score(X_test,Y_test)
rf_score
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import gradio as gr
import joblib

# Load the saved Random Forest model
rf = joblib.load('random_forest_model.joblib')

# Load the saved LabelEncoder map
le_map = joblib.load('label_encoder_map.joblib')

# Reload the dataset to get original categorical values for Gradio dropdowns
df_raw = pd.read_csv("/content/laptops.csv")

# Handle missing values (same as in the notebook's preprocessing)
df_raw["rating"] = df_raw["rating"].fillna(df_raw["rating"].median())
df_raw["no_of_ratings"] = df_raw["no_of_ratings"].fillna(df_raw["no_of_ratings"].median())
df_raw["no_of_reviews"] = df_raw["no_of_reviews"].fillna(df_raw["no_of_reviews"].median())

# Drop irrelevant columns (same as in the notebook's preprocessing)
df_raw = df_raw.drop(["Unnamed: 0", "img_link"], axis=1, errors="ignore")

# Define the prediction function for Gradio
def predict_price(name_val, processor_val, ram_val, os_val, storage_val, display_size_val, rating_val, no_of_ratings_val, no_of_reviews_val):
    # Encode categorical inputs using the loaded LabelEncoders
    encoded_name = le_map['name'].transform([name_val])[0]
    encoded_processor = le_map['processor'].transform([processor_val])[0]
    encoded_ram = le_map['ram'].transform([ram_val])[0]
    encoded_os = le_map['os'].transform([os_val])[0]
    encoded_storage = le_map['storage'].transform([storage_val])[0]

    # Create a DataFrame for the new input, ensuring column order matches X used during training
    feature_columns = ['name', 'processor', 'ram', 'os', 'storage', 'display(in inch)', 'rating', 'no_of_ratings', 'no_of_reviews']
    input_data = [
        encoded_name,
        encoded_processor,
        encoded_ram,
        encoded_os,
        encoded_storage,
        display_size_val,
        rating_val,
        no_of_ratings_val,
        no_of_reviews_val
    ]
    input_df = pd.DataFrame([input_data], columns=feature_columns)

    # Predict using the Random Forest model
    prediction = rf.predict(input_df)[0]
    return f"Predicted Laptop Price: Rs. {prediction:,.2f}"

# Get unique values for dropdowns from the raw data for interface choices
name_choices = sorted(df_raw['name'].unique().tolist())
processor_choices = sorted(df_raw['processor'].unique().tolist())
ram_choices = sorted(df_raw['ram'].unique().tolist())
os_choices = sorted(df_raw['os'].unique().tolist())
storage_choices = sorted(df_raw['storage'].unique().tolist())

# Redesigning the Gradio interface using gr.Blocks for more control and a polished look
# Removed 'mode='dark'' as it caused a TypeError
with gr.Blocks(theme=gr.themes.Default(primary_hue='orange'), title="Laptop Price Predictor") as demo:
    gr.Markdown(
        """
        # Laptop Price Predictor
        Enter the specifications of a laptop to get an estimated price.
        This model uses a Random Forest Regressor and has an R2 score of approximately 87%.
        """
    )
    with gr.Row():
        with gr.Column():
            name_input = gr.Dropdown(name_choices, label="Laptop Name", value=name_choices[0])
            processor_input = gr.Dropdown(processor_choices, label="Processor Type", value=processor_choices[0])
            ram_input = gr.Dropdown(ram_choices, label="RAM Configuration", value=ram_choices[0])
        with gr.Column():
            os_input = gr.Dropdown(os_choices, label="Operating System", value=os_choices[0])
            storage_input = gr.Dropdown(storage_choices, label="Storage Type", value=storage_choices[0])
            display_input = gr.Number(label="Display Size (in inch)", value=15.6)
    with gr.Row():
        rating_input = gr.Slider(minimum=1.0, maximum=5.0, step=0.1, label="Rating", value=4.3)
        num_ratings_input = gr.Number(label="Number of Ratings", value=90)
        num_reviews_input = gr.Number(label="Number of Reviews", value=11)

    predict_button = gr.Button("Predict Price")
    output_textbox = gr.Textbox(label="Predicted Price", interactive=False)

    predict_button.click(
        fn=predict_price,
        inputs=[
            name_input,
            processor_input,
            ram_input,
            os_input,
            storage_input,
            display_input,
            rating_input,
            num_ratings_input,
            num_reviews_input
        ],
        outputs=output_textbox
    )

demo.launch(share=True)
```
